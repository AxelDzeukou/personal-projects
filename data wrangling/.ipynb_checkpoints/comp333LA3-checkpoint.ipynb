```python
#40089940
#Axel Dzeukou
#google drive link to final output csv https://drive.google.com/file/d/119lrTo9uoEyDeStF_lXRsk80VzxRRY9q/view?usp=sharing
import nltk
import pandas as pd
from nltk.tokenize import WhitespaceTokenizer
import string
import math
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

import unicodedata
def remove_control_characters(s):
    return "".join(ch for ch in s if unicodedata.category(ch)[0]!="C")

def generate_N_grams(text,ngram=1):
  words=[word for word in text.split(" ")]
#   print("Sentence after removing stopwords:",words)
  temp=zip(*[words[i:] for i in range(0,ngram)])
  ans=[' '.join(ngram) for ngram in temp]
  return ans

#text facet function
def textfacet(df,column):
    return df[column].value_counts()

#numeric facet
def numericfacet(df,column,include_set):
#create histogram and decide to include numeric,non-numeric, blank and error. anything non-numeric is 0

#number of numeric
 print('numeric: '+ str(len(df[column].dropna().str.isnumeric())))
 
#number of non-numeric
 print('non-numeric: '+ str(len(df[column])-len(df[column].dropna().str.isnumeric())))

#number of blank
 print('blank: '+ str(df[column].isna().sum()))

 histo_data=[]   
 
 df['numeric'] = df[column].apply(np.isreal)

#replace extra columns with zero
 for field in include_set:
  if field=='non-numeric':
   df.loc[df['numeric'] == 0, column] = 0

  if field=='blank':
   df[column] = df[column].fillna(0)

 #errors='ignore'
 pd.to_numeric(df[column], downcast='float')
 df[column].apply(lambda row: histo_data.append(row))
 

#plot histo
 plt.hist(histo_data)

#TypeError: cannot perform reduce with flexible type??
 print('min-max')
 print(stats.describe(histo_data)[1])

#text clustering
def text_clustering(df,column):
 # Convert the dictionary into DataFrame

    #change all characters to their lowercase representation
    df['ngram'] = df[column].apply(lambda row: row.lower())

    #omit punctuation
    df['ngram'] = df[column].apply(lambda row: row.translate(str.maketrans('', '', string.punctuation)))

    #remove control characters
    df['ngram'] = df[column].apply(lambda row: remove_control_characters(row))

    #remove whitespace
    df['ngram'] = df[column].apply(lambda row: WhitespaceTokenizer().tokenize(row))
    df['ngram'] = df[column].apply(lambda row: "".join(row))

    #sort and remove duplicate n-grmas,join
    df['ngram'] = df[column].apply(lambda row: "".join(list(sorted(set(generate_N_grams(row,ngram=2))))))

    #normalize extended western characters to their ASCII representation??
    
    #group rows based on terms
    grouped_df = df.groupby("ngram")

    #create a list of tokenids on each grouped term as postings list
    grouped_lists = grouped_df[column].apply(list)
    
    print(display(grouped_lists))
    
    return df['ngram']

#input df
df=pd.read_csv(r"C:\Users\Axel\Downloads\Portal_rodents_19772002_scinameUUIDs (2).csv",encoding='windows-1252')

#text facet of column scientificName
print('text facet of column scientificName')
print(textfacet(df,'scientificName'))

#convert recordID,period,and stake to number field
fields=['recordID','period','stake']

#dataframe for numeric facet and scatterplot
numeric_df=df.copy()
for field in fields:
 numeric_df[field] = numeric_df[field].fillna(0).apply(int)

#numeric facet of period reverse after
#'NaN' is actually just a string
numeric_df.loc[numeric_df["period"] == 90, "period"] = 'hey'
numeric_df.loc[numeric_df["period"] == 100, "period"] = math.nan

print("numeric facet of column period")
numericfacet(numeric_df,'period',['numeric','non-numeric', 'blank'])

#scatterplot
fig, axs = plt.subplots(1,len(fields)-1, figsize=(30, 30))
fig.suptitle('square grid of scatterplots ', fontsize=30)

for i in range(0,len(fields[1:])):
    axs[i].set_title("scatterplot of "+fields[0]+' and '+fields[i+1])
    axs[i].scatter(numeric_df[fields[0]],numeric_df[fields[i+1]])
    
#select plot you want
# axs[1]

# axs[2]

#perform n-gram text cluster then merge?? for now removed rows with nan
df['scientificName'] = df['scientificName'].fillna('NaN')
df=df.loc[df['scientificName'] !='NaN']
df['scientificName']=text_clustering(df,'scientificName')

df=df.drop(['ngram'], axis=1)
```

```python
#Trimming Leading and Trailing Whitespace on unsplittered data

df['scientificName'] = df['scientificName'].apply(lambda row: row.lstrip())
df['scientificName'] = df['scientificName'].apply(lambda row: row.rstrip())

#replace middle whitespaces with space
df['scientificName'] = df['scientificName'].apply(lambda row: WhitespaceTokenizer().tokenize(row))
df['scientificName'] = df['scientificName'].apply(lambda row: " ".join(row))
 

#data splitting clustered data on column scientificName
df[['scientificName 1', 'scientificName 2']] = df['scientificName'].str.split(' ',expand=True)

    
#Renaming columns
df.rename(columns = {'scientificName 1':'genus', 'scientificName 2':'species'}, inplace = True)

#save modified data to csv
df.to_csv(r"C:\Users\Axel\Desktop\newwine\comp333\labAssignment3\out.csv") 
```
